```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
from matplotlib.pyplot import title
```

```python
data_frame = pd.read_csv("./Employee Sample Data.csv")
data_frame
```

```python
males = data_frame[data_frame["Gender"] == "Male"]
males
```

```python
females = data_frame[data_frame["Gender"] == "Male"]
females
```

```python
male_ages = males["Age"]
male_ages
```

```python
female_ages = females["Age"]
female_ages
```

```python
plt.xlabel("ages")
plt.ylabel("employees count")
plt.title("employees count by age")
plt.plot(male_ages,np.arange(len(male_ages)))
plt.legend("males")
plt.plot(female_ages,np.arange(len(female_ages)))
plt.legend("females")
plt.grid(True)
```

```python
female_ages.plot(xlabel="ages",ylabel="count",title="males",figsize=(15,10))
male_ages.plot(xlabel="ages",ylabel="count",title="females",figsize=(15,10))
```
