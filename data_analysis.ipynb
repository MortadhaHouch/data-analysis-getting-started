```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("./Employee Sample Data.csv")
data_frame
```

```python
males = data_frame[data_frame["Gender"] == "Male"]
males.head(5)
```

```python
females = data_frame[data_frame["Gender"] == "Female"]
females.groupby("Department").head(5)
```

```python
males_still_working = males[males["Exit Date"].isna()].groupby("Hire Date")
males_still_working.head(10).plot(xlabel="count",ylabel="age",title="progress of man unemployment quitting based on age",figsize=(15,5))
```

```python
females_still_working = females[females["Exit Date"].isna()].groupby("Hire Date")
females_still_working.head(10).plot(xlabel="count",ylabel="age",title="progress of female unemployment quitting based on age",figsize=(15,5))
```

```python
males_grouped_by_country = males.groupby("Country")
males_grouped_by_country.plot(xlabel="age",ylabel="count",title="males count based on country",figsize=(15,5))
```

```python
females_grouped_by_country = females.groupby("Country")
females_grouped_by_country.plot(xlabel="females",ylabel="age",title="females count based on country",figsize=(15,5))
```

```python
females.plot(xlabel="count",ylabel="age",title="females count per age",figsize=(15,5))
males.plot(xlabel="count",ylabel="age",title="males count per age",figsize=(15,5))
```

```python
males_salaries = males["Annual Salary"].str.replace(r"[$,]","",regex=True).astype(float)/12
males_salaries
```

```python
females_salaries = females["Annual Salary"].str.replace(r"[$,]","",regex=True).astype(float)/12
females_salaries
```

```python
plt.title("males salaries compared to females salaries")
plt.figure(figsize=(20,5))
plt.xlabel("Annual Salary")
plt.ylabel("count")
plt.plot(np.arange(100),males_salaries[:100],label="males salaries",linestyle="dashed")
plt.plot(np.arange(100),females_salaries[:100],label="males salaries",linestyle="dashed")
plt.grid(True)
plt.legend(["males","females"])
plt.show()
```

```python
males_paid_more_females = 0
females_paid_more_males = 0
females_paid_equally_with_males = 0
for i in range(0,400):
    if np.array(males_salaries)[i] > np.array(females_salaries)[i]:
        males_paid_more_females+=1
    elif np.array(males_salaries)[i] < np.array(females_salaries)[i]:
        females_paid_more_males+=1
    elif np.array(females_salaries)[i] == np.array(males_salaries)[i]:
        females_paid_equally_with_males+=1
print(males_paid_more_females,females_paid_more_males,females_paid_equally_with_males)
```
