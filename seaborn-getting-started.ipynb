```python
import seaborn as sns
```

```python
data_sets = sns.get_dataset_names()
data_sets
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
sns.pointplot(x="price",y="carat",data=diamonds.head(10))
```

```python
sns.scatterplot(x="price", y="carat", data=diamonds.head(10),hue="carat",size="carat")
```

```python
sns.barplot(x="price", y="carat", data=diamonds.head(10), hue="carat")
```

```python
sns.histplot(x="price", y="carat", data=diamonds.head(10), hue="carat")
```

```python
sns.boxplot(x="price", y="carat", data=diamonds.head(10), hue="carat")
```

```python
sns.jointplot(x="price", y="carat", data=diamonds.head(10), hue="carat")
```

```python
sns.stripplot(x="price", y="carat", data=diamonds.head(10), hue="carat",dodge=True)
```

```python
sns.boxplot(x="price", y="carat", data=diamonds.head(10), hue="carat",dodge=True)
```

```python
sns.displot(x="price", y="carat", data=diamonds.head(10), hue="carat",kind="hist")
```

```python
sns.jointplot(x="price", y="carat", data=diamonds.head(10),kind="reg")
```

```python
sns.jointplot(x="price", y="carat", data=diamonds.head(10),kind="hex")
```
