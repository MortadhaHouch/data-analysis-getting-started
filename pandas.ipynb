```python
annual_salaries = data_frame["Annual Salary"]
annual_salaries
```

```python
numeric_annual_salaries = pd.to_numeric(annual_salaries.str.replace(r"[$,]","",regex=True).astype(float))
numeric_annual_salaries
```

```python
numeric_annual_salaries.plot(kind="bar",figsize=(10,5),title="Annual Salary",xlabel="count",ylabel="Annual Salary")
```

```python
males_salaries = data_frame[data_frame["Gender"]=="Male"]["Annual Salary"].str.replace(r"[$,]","",regex=True).astype(float)
females_salaries = data_frame[data_frame["Gender"]=="Female"]["Annual Salary"].str.replace(r"[$,]","",regex=True).astype(float)
males_salaries.plot(xlabel="count (person)",ylabel="yearly salary ($)",title="salary per count",figsize=(15,10))
females_salaries.plot(xlabel="count (person)",ylabel="yearly salary ($)",title="salary per count",figsize=(15,10))
```

```python
males_salaries = data_frame[data_frame["Gender"] == "Male"]["Annual Salary"].str.replace(r"[$,]", "",regex=True).astype(float)
females_salaries = data_frame[data_frame["Gender"] == "Female"]["Annual Salary"].str.replace(r"[$,]", "",regex=True).astype(float)
males_salaries/=12
females_salaries/=12
males_salaries.sample(20).plot(xlabel="count (person)", ylabel="monthly salary ($)", title="salary per count",figsize=(15,10),kind="line")
females_salaries.sample(20).plot(xlabel="count (person)", ylabel="monthly salary ($)", title="salary per count",figsize=(15,10),kind="line")
```

```python
plt.xlabel("salaries")
plt.ylabel("count")
plt.title("salary per count")
plt.plot(males_salaries[0:100],np.arange(100),label="male salaries",linestyle="dashed")
plt.plot(females_salaries[0:100],np.arange(100),label="females salaries",linestyle="dashed")
plt.legend()
plt.grid()
```

```python
plt.xlabel("salaries")
plt.ylabel("count")
plt.title("salary per count")
plt.hist(males_salaries[0:100],np.arange(100),label="male salaries",linestyle="dashed")
plt.hist(females_salaries[0:100],np.arange(100),label="females salaries",linestyle="dashed")
plt.legend()
plt.grid()
```

```python
plt.xlabel("")
plt.ylabel("")
plt.title("salary per count")
plt.pie(males_salaries[0:100],labels=np.arange(100))
plt.pie(females_salaries[0:100],labels=np.arange(100))
plt.legend()
plt.grid()
```
